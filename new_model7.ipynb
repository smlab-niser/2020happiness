# **COMPARATIVE PLOTS FOR ML ALGORITHMS(LR, SVM, KNN, RF) AND CLUSTERING BASED ON HAPPINESS CLASSES USING THE UTKAL UNIVERSITY TEST DATA.**




```python
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
%matplotlib inline
```

```python
cols = ['Freedom', 'Social', 'HLE']
```

```python
happy = pd.read_csv('testdata.csv')
```

##**Comparative plot of 4 ML algorithms for happiness classes and prdicted variables(social, freedom, health)

```python



sns.pairplot(happy[cols], hue='Positive_Group')
plt.show()
```

```python
cols.append('Positive_Group')
```


##  **USING THE HAPPINESS WORLD DATASET, SHOWING A MATRIX COMPARATIVE PLOT FOR PREDICTED PARAMETERS (FREEDOM SOCIAL AND HEALTH) USING THE ML ALGORTITHMS(KNN RF SVN LR) AND CLUSTERING**

```python
plt.figure(figsize=(30,25))

sns.lmplot(data=happy, x='Happiness Class', y='LR+Clustering', hue='Positive_Group')
plt.xlabel('Happiness Class'),plt.ylabel('LR+Clustering')
sns.lmplot(data=happy, x='Happiness Class', y='KNN+Clustering', hue='Positive_Group')
plt.xlabel('Happiness Class'), plt.ylabel('KNN+Clustering')
sns.lmplot(data=happy, x='Happiness Class', y='HLE', hue='SVM+Clustering')
plt.xlabel('Happiness Class'),plt.ylabel('SVM+Clustering')

plt.show()
```
